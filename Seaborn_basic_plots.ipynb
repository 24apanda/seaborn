# Pair Grid

```python
iris = sns.load_dataset('iris')
```

```python
iris.head(5)
```

```python
x = sns.PairGrid(iris)
x = x.map(plt.scatter)
```

```python
x = sns.PairGrid(iris)
x = x.map_diag(plt.hist)
x = x.map_offdiag(plt.scatter)
```

```python
x = sns.PairGrid(iris,hue ='species')
x = x.map_diag(plt.hist)
x = x.map_offdiag(plt.scatter)
```

```python
x = sns.PairGrid(iris,hue ='species')
x = x.map_diag(plt.hist)
x = x.map_offdiag(plt.scatter)
x = x.add_legend()
```

```python
x = sns.PairGrid(iris,hue ='species')
x = x.map(plt.scatter)
```

```python
x = sns.PairGrid(iris,hue ='species',vars =['petal_length','petal_width'])
x = x.map(plt.scatter)
```

```python
x = sns.PairGrid(iris)
x = x.map_diag(plt.hist)
x = x.map_upper(plt.scatter)
x = x.map_lower(sns.kdeplot)
x = x.add_legend()
```

# KDE Plot

```python
mean =[0,0]
cov = [[0.2,0],[0,3]]
x_axis,y_axis = np.random.multivariate_normal(mean,cov,size=40).T
```

```python
sns.kdeplot(x_axis)
```

```python
sns.kdeplot(x_axis,y_axis)
```

```python
sns.kdeplot(x_axis,shade=True)
```

# Reg plot

```python
sns.regplot(x='total_bill',y='tip',data=tips)
```

```python
mean =[2,5]
cov = [[1.1,0.4],[2.2,3]]
x_value,y_value = np.random.multivariate_normal(mean,cov,size=40).T
```

```python
sns.regplot(x=x_value,y=y_value)
```

```python
sns.regplot(x='total_bill',y='tip',data=tips,scatter_kws={'marker':'D','color':'blue'},line_kws={'color':'red','linewidth':3.1})
```

```python
sns.regplot(x='total_bill',y='tip',data = tips,x_estimator = np.mean)
```
